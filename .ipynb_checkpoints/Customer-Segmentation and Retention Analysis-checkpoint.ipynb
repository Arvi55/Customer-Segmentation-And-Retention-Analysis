```python
#Recency = Reference Date - Customer’s Last Purchase Date
```

```python
reference_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
```

```python
reference_date
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalPrice': 'sum'
})
```

```python
rfm
```

```python
rfm.columns = ['Recency', 'Frequency', 'Monetary']
```

```python
rfm
```

```python
rfm.describe()
```

```python
(rfm["Monetary"]<= 0).sum()
```

```python
#i am removing these 50 customer who have 0 or less monetary because they do not contribute in any way and business dont need them
rfm = rfm[rfm['Monetary'] > 0]
rfm.shape
```

```python
#next i am building a scoring system(1-5) for R,F,M based on quantile 
#i mean, i will be spliting the customer on the scale form 1 to 5 ok 
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 
                         q=5, 
                         labels=[5,4,3,2,1])
```

```python
rfm
```

```python
rfm[['Recency', 'R_Score']].head()
```

```python
#to see the destribution of score in our data 
rfm['R_Score'].value_counts()
```

```python
rfm['F_Score'] = pd.qcut(
    rfm['Frequency'],
    q=5,
    duplicates='drop'
)
rfm['F_Score'] = rfm['F_Score'].cat.codes + 1

```

```python
rfm['F_Score'].value_counts()
```

```python
rfm['M_Score']= pd.qcut(rfm['Monetary'],
                       q=5,
                       labels= [1,2,3,4,5])
rfm['M_Score'].value_counts()
```

```python
#now i am adding the R,F,M score to make one string( eg:-555,541 etc)
rfm['RFM_Score'] = (
    rfm['R_Score'].astype(str) +
    rfm['F_Score'].astype(str) +
    rfm['M_Score'].astype(str)
)
```

```python
rfm[['R_Score','F_Score','M_Score','RFM_Score']].head()
```

```python
rfm["RFM_Score"].nunique()
```

```python
# now i will dividing the customer into diffrent diffrent segment such as loyal , champions, big spender etc
def segment_customer(row):
    if row['R_Score'] >= 4 and row['F_Score'] >= 3 and row['M_Score'] >= 4:
        return 'Champions'
    elif row['F_Score'] >= 3 and row['R_Score'] >= 3:
        return 'Loyal Customers'
    elif row['R_Score'] >= 4 and row['F_Score'] == 2:
        return 'Potential Loyalists'
    elif row['M_Score'] >= 4 and row['F_Score'] <= 2:
        return 'Big Spenders'
    elif row['R_Score'] <= 2 and row['F_Score'] >= 3:
        return 'At Risk'
    elif row['R_Score'] <= 2 and row['F_Score'] <= 2 and row['M_Score'] <= 2:
        return 'Hibernating'
    elif row['R_Score'] == 1 and row['M_Score'] >= 4:
        return 'Lost High Value'
    else:
        return 'Others'
```

```python
rfm['Segment'] = rfm.apply(segment_customer, axis=1)
```

```python
rfm['Segment'].value_counts()
```

```python
segment_revenue = rfm.groupby('Segment')['Monetary'].sum().sort_values(ascending=False)
segment_revenue
```

```python
rfm['Recency'].describe()
```

```python
rfm['Churn'] = (rfm['Recency'] > 90).astype(int)
rfm['Churn'].value_counts()

```

```python
#This will show churn rate per segment.
#This answers:
#Are “Hibernating” really churned?
#Are “Champions” safe?

Is “At Risk” behaving correctly?
churn_by_segment = (rfm.groupby('Segment')['Churn'].mean().sort_values(ascending=False)) *100
churn_by_segment

```

```python
#If we do nothing, how much revenue do we lose?
revenue_at_risk = rfm[rfm['Churn'] == 1]['Monetary'].sum()
total_revenue = rfm['Monetary'].sum()

revenue_at_risk, total_revenue
```

```python
# so about 10% of revenue is at RISK if we dont do anything ummmmm...

#ok so Out of ₹8.5 lakh at risk,How much belongs to Big Spenders & At Risk segments?
rfm[rfm['Churn'] == 1].groupby('Segment')['Monetary'].sum().sort_values(ascending=False)
```
